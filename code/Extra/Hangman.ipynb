```python
# hangman problem


import random

# List of words to use for the game
word_list = ["python", "programming", "computer", "science", "hangman", "game", "development"]


def get_random_word(word_list):
    # Returns a randomly chosen word from a list of words
    return random.choice(word_list)


def display_word(word, guessed_letters):
    # Returns the current state of the word to display to the user
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter
        else:
            display += "_"
    return display


def play_game(word):
    # Runs the main game loop
    guessed_letters = set()
    remaining_guesses = 6
    game_over = False  # Flag variable to track game state
    while remaining_guesses > 0 and not game_over:
        print(f"Word: {display_word(word, guessed_letters)}")
        print(f"Remaining guesses: {remaining_guesses}")
        guess = input("Guess a letter: ").lower()
        if guess in guessed_letters:
            print("You've already guessed that letter.")
        elif guess in word:
            print("Correct!")
            guessed_letters.add(guess)
            if set(word) <= guessed_letters:
                print("Congratulations, you've won!")
                game_over = True  # Set game_over flag to True
        else:
            print("Incorrect.")
            guessed_letters.add(guess)
            remaining_guesses -= 1
    if not game_over:  # Check if the game ended without winning
        print(f"Sorry, you've lost. The word was {word}.")


# Main program loop
while True:
    print("Welcome to Hangman!")
    word = get_random_word(word_list)
    play_game(word)
    play_again = input("Would you like to play again? (y/n): ").lower()
    if play_again != "y":
        break
```
